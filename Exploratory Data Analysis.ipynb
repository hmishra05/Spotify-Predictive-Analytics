````python
# lib imports
import pandas as pd
import csv
from datetime import datetime 
````

````python
data_location = "Data/Spotify_Data_Sample.csv"
````

## Initial Data Cleaning

````python
df = pd.read_csv(data_location)
df.head()
````

````python
desired_columns = ['name',
 'collaborative',
 'pid',
 'modified_at',
 'num_albums',
 'num_tracks',
 'num_edits',
 'duration_ms',
 'num_artists',
 'num_followers']

````

````python
df
````

````python
df = df[desired_columns]
````

Data Description
```
name: string

pid: Number

num_albums: Number

num_artist: Number

```

## Data Cleaning

````python
# Initial types
df.dtypes
````

````python
# Type casting the columns
df['name'] = df['name'].astype('str')
df['collaborative'] = df['collaborative'].astype('bool')
df['pid'] = df['pid'].astype('int')

# Date conversions
df['modified_at'] = pd.to_datetime(df['modified_at'], unit='s')
df['year'] = df['modified_at'].dt.year
df['month'] = df['modified_at'].dt.month

# Other numeric conditions
df['num_albums'] = df['num_albums'].astype("int")
df['num_tracks'] = df['num_tracks'].astype("int")

# Will impute the non numeric values to NaN using coerce
df['num_edits'] = pd.to_numeric(df['num_edits'], errors='coerce')
df['duration_ms'] = pd.to_numeric(df['duration_ms'], errors='coerce')
df['num_artists'] = pd.to_numeric(df['num_artists'], errors='coerce')
df['num_followers'] = pd.to_numeric(df['num_followers'], errors='coerce')
````

````python
# Dropping the column for redundant info
print(df.columns)
df = df.drop(['modified_at'], axis=1)
print(df.columns)
````
